# Introduction

Cette leçon introduit la **théorie du consommateur**, qui cherche à modéliser les choix d’un individu confronté à une contrainte de revenu et plusieurs combinaisons possibles de biens à consommer.

# 2. Contrainte budgétaire

Le consommateur a un revenu $R > 0$ et fait face à des prix $p_1$ et $p_2$.

La contrainte budgétaire est donnée par :

$$
p_1 x_1 + p_2 x_2 \leq R
$$

L’ensemble des paniers accessibles (ensemble budgétaire) est :

$$
B = \left\{ (x_1, x_2) \in \mathbb{R}_+^2 \, | \, p_1 x_1 + p_2 x_2 \leq R \right\}
$$

> 💡 La droite budgétaire correspond à l’égalité $p_1 x_1 + p_2 x_2 = R$.

````


⸻

3. Maximisation de l’utilité

Le consommateur choisit le panier $(x_1^, x_2^)$ qui maximise son utilité sous la contrainte budgétaire.

3.1 Programme du consommateur

$$
\max_{x_1, x_2} u(x_1, x_2) \quad \text{sous la contrainte} \quad p_1 x_1 + p_2 x_2 \leq R
$$

3.2 Méthode du Lagrangien

Définissons le Lagrangien :

$$
\mathcal{L}(x_1, x_2, \lambda) = u(x_1, x_2) - \lambda , (p_1 x_1 + p_2 x_2 - R)
$$

Les conditions du premier ordre (CPO) sont :

$$
\begin{aligned}
\frac{\partial \mathcal{L}}{\partial x_1} &= \frac{\partial u}{\partial x_1} - \lambda p_1 = 0 \quad \Rightarrow \quad \frac{\partial u}{\partial x_1} = \lambda p_1, \\
\frac{\partial \mathcal{L}}{\partial x_2} &= \frac{\partial u}{\partial x_2} - \lambda p_2 = 0 \quad \Rightarrow \quad \frac{\partial u}{\partial x_2} = \lambda p_2, \\
\frac{\partial \mathcal{L}}{\partial \lambda} &= R - p_1 x_1 - p_2 x_2 = 0 \quad \Rightarrow \quad p_1 x_1 + p_2 x_2 = R.
\end{aligned}
$$

La condition d’optimalité qui en découle est :

$$
\frac{\frac{\partial u}{\partial x_1}}{\frac{\partial u}{\partial x_2}} = \frac{p_1}{p_2},
$$

ce qui signifie que le taux marginal de substitution (TMS) doit être égal au rapport des prix.

⸻

4. Exemples types

4.1 Fonction Cobb-Douglas

Soit $u(x_1, x_2) = x_1^\alpha x_2^{1-\alpha}$ avec $0 < \alpha < 1$.

Demande optimale :

$$
x_1^* = \alpha \frac{R}{p_1}, \quad x_2^* = (1 - \alpha) \frac{R}{p_2}.
$$

Démonstration détaillée :
	1.	Formulation du Lagrangien :

$$
\mathcal{L}(x_1, x_2, \lambda) = x_1^\alpha x_2^{1-\alpha} - \lambda (p_1 x_1 + p_2 x_2 - R).
$$
	2.	Conditions du premier ordre :

$$
\begin{aligned}
\frac{\partial \mathcal{L}}{\partial x_1} &= \alpha x_1^{\alpha-1} x_2^{1-\alpha} - \lambda p_1 = 0, \\
\frac{\partial \mathcal{L}}{\partial x_2} &= (1-\alpha)x_1^\alpha x_2^{-\alpha} - \lambda p_2 = 0, \\
\frac{\partial \mathcal{L}}{\partial \lambda} &= R - p_1 x_1 - p_2 x_2 = 0.
\end{aligned}
$$
	3.	En divisant les deux premières équations pour éliminer $\lambda$ :

$$
\frac{\alpha x_1^{\alpha-1} x_2^{1-\alpha}}{(1-\alpha)x_1^\alpha x_2^{-\alpha}} = \frac{p_1}{p_2} \quad \Rightarrow \quad \frac{\alpha}{1-\alpha} \cdot \frac{x_2}{x_1} = \frac{p_1}{p_2}.
$$
	4.	On obtient :

$$
\frac{x_2}{x_1} = \frac{1-\alpha}{\alpha} \cdot \frac{p_1}{p_2}.
$$
	5.	En substituant dans la contrainte budgétaire, on en déduit :

$$
x_1^* = \alpha \frac{R}{p_1} \quad \text{et} \quad x_2^* = (1-\alpha) \frac{R}{p_2}.
$$

4.2 Biens parfaits substituts

Pour une fonction d’utilité du type :

$$
u(x_1, x_2) = a x_1 + b x_2,
$$

le consommateur achètera uniquement le bien ayant le meilleur rapport utilité-prix, c’est-à-dire en comparant $\frac{a}{p_1}$ et $\frac{b}{p_2}$.

4.3 Biens parfaits compléments

Pour une fonction d’utilité :

$$
u(x_1, x_2) = \min{a x_1, b x_2},
$$

la demande optimale se détermine en imposant :

$$
a x_1 = b x_2 \quad \text{et} \quad p_1 x_1 + p_2 x_2 = R.
$$

⸻

5. Exercices corrigés

Exercice 1 — Fonction Cobb-Douglas

Soit $u(x_1, x_2) = x_1^{0.5} x_2^{0.5}$, $R = 100$, $p_1 = 2$, $p_2 = 5$.

Calculs :
	•	Ici, $\alpha = 0.5$.
	•	$x_1^* = 0.5 \cdot \frac{100}{2} = 25$.
	•	$x_2^* = 0.5 \cdot \frac{100}{5} = 10$.

Réponse : $x_1^* = 25$, $x_2^* = 10$.

Exercice 2 — Biens parfaits compléments

Soit $u(x_1, x_2) = \min{x_1, 2x_2}$, $R = 120$, $p_1 = 4$, $p_2 = 6$.

Étapes de résolution :
	1.	Condition de complémentarité : pour maximiser l’utilité, il faut que $x_1 = 2x_2$.
	2.	Substitution dans la contrainte budgétaire :

$$
4(2x_2) + 6x_2 = 8x_2 + 6x_2 = 14x_2 = 120.
$$
	3.	Résolution :

$$
x_2^* = \frac{120}{14} = \frac{60}{7}, \quad x_1^* = 2 \cdot \frac{60}{7} = \frac{120}{7}.
$$

Réponse : $x_1^* = \frac{120}{7}$, $x_2^* = \frac{60}{7}$.
